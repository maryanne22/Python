```python
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['TotalPay'].fillna(df['TotalPay'].mean(), inplace=True)
df['TotalPayBenefits'].fillna(df['TotalPayBenefits'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

## drop columns

```python
columns_to_drop = ['Notes', 'Agency', 'Status']
df.drop(columns=columns_to_drop, axis=1, inplace=True)


```

## 2. Find occurrence of employee names

```python
employee_name_counts = df['EmployeeName'].value_counts()
employee_name_counts
```

## 3. Find number of unique job titles

```python
unique_job_titles = df['JobTitle'].nunique()
unique_job_titles
```

## 4. Display all the employee names from fire department

```python
fire_department_employees = df[df['JobTitle'].str.contains('Fire', case=False)]

fire_department_employee_names = fire_department_employees['EmployeeName']
print(fire_department_employee_names)
```

## 5. Find the minimum, maximum and average basepay

```python

min_base_pay = df['BasePay'].min()
max_base_pay = df['BasePay'].max()
average_base_pay = df['BasePay'].mean()

print("Minimum BasePay:", min_base_pay)
print("Maximum BasePay:", max_base_pay)
print("Average BasePay:", average_base_pay)
```

## 6. Replace "Not provided" in employeename column with NAN

```python
df['EmployeeName'] = df['EmployeeName'].replace("Not provided", np.nan)
```

```python
df
```

## 7. Drop rows having 5missing values

```python
threshold = len(df.columns)- 5
df.dropna(thresh=threshold, inplace=True)
```

## 8. Find the job title of ALBERT PARDINI

```python
albert_pardini_job_title = df[df['EmployeeName'] == 'ALBERT PARDINI']['JobTitle']

print("Job Title of ALBERT PARDINI:", albert_pardini_job_title.values[0])

```

## 9. How much he makes including benefits

```python
albert_pardini_total_pay_benefits = df[df['EmployeeName'] == 'ALBERT PARDINI']['TotalPayBenefits']
print("Total Pay and Benefits for ALBERT PARDINI:", albert_pardini_total_pay_benefits.values[0])
```

## 10.  Show the name of the person with highest basepay

```python
index_of_highest_base_pay = df['BasePay'].idxmax()
person_with_highest_base_pay = df.loc[index_of_highest_base_pay, 'EmployeeName']
print("Name of the person with the highest BasePay:", person_with_highest_base_pay)
```

## 11.  Find the average basepay of all employees per job title

```python
average_basepay_per_job_title = df.groupby('JobTitle')['BasePay'].mean()
print(average_basepay_per_job_title)
```

## 12.  Find the average basepay of all employees having job title Accountant

```python
accountants = df[df['JobTitle'] == 'Accountant']
average_basepay_accountant = accountants['BasePay'].mean()
print("Average BasePay for Accountants:", average_basepay_accountant)
```

## 13. Find the five most common jobs

```python
job_title_counts = df['JobTitle'].value_counts()
top_five_common_jobs = job_title_counts.head(5)
print("Five Most Common Job Titles:")
print(top_five_common_jobs)
```
